# Speedup

````python
# https://pandas.pydata.org/pandas-docs/stable/reference/series.html#conversion
def speedup(df, mpimin):
    # tzero = float(df[ df.mpi.isin(['0036mpi']) ].internal)
    tzero = float(df[ df.mpi.isin([mpimin]) ].internal)
    mmzero = int( mpimin.strip('mpi') )
    nzero = int(df[ df.mpi.isin([mpimin]) ].n_cube)
    ncube = nzero ** 3
    for tn in df.internal:    
        mm = int(  df[df.internal.isin([tn])].mpi.to_list()[0].strip('mpi') )
        npc = int(ncube / mm)
        ss = round(tzero / tn, 3)
        efficiency = round(100 * ss / mm  * mmzero, 1)
        print('n=%s ncube=%s npc=%s mpi=%s tn=%s speedup=%s //eff=%s' % (nzero, ncube, npc, mm, tn, ss, efficiency))
    return 0
````
