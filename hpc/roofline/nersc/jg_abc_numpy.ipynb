```python
dfjg = df.copy()[columns_l][1:]
dfjg
```

```python
for kk, vv in dfjg.iteritems():   
    if kk == 'sm__cycles_elapsed.avg': cyc_avg = vv.tolist()
    elif kk == 'sm__cycles_elapsed.avg.per_second': cyc_avg_per_second = vv.tolist()    
    elif kk == 'sm__sass_thread_inst_executed_op_dmul_pred_on.sum': dmul = vv.tolist()
    elif kk == 'sm__sass_thread_inst_executed_op_dadd_pred_on.sum': dadd = vv.tolist()
    elif kk == 'sm__sass_thread_inst_executed_op_dfma_pred_on.sum': dfma = vv.tolist()
    elif kk == 'sm__sass_thread_inst_executed_op_fmul_pred_on.sum': fmul = vv.tolist()
    elif kk == 'sm__sass_thread_inst_executed_op_fadd_pred_on.sum': fadd = vv.tolist()
    elif kk == 'sm__sass_thread_inst_executed_op_ffma_pred_on.sum': ffma = vv.tolist()
    elif kk == 'sm__sass_thread_inst_executed_op_hmul_pred_on.sum': hmul = vv.tolist()
    elif kk == 'sm__sass_thread_inst_executed_op_hadd_pred_on.sum': hadd = vv.tolist()
    elif kk == 'sm__sass_thread_inst_executed_op_hfma_pred_on.sum': hfma = vv.tolist()
    elif kk == 'sm__inst_executed_pipe_tensor.sum': tensor_inst = vv.tolist()
    elif kk == 'dram__bytes.sum': dram = vv.tolist()
    elif kk == 'lts__t_bytes.sum': l2 = vv.tolist()
    elif kk == 'l1tex__t_bytes.sum': l1 = vv.tolist()      
    # peaks:
    elif kk == 'sm__sass_thread_inst_executed_op_ffma_pred_on.sum.peak_sustained': peak_ffma = vv.tolist()
    elif kk == 'sm__sass_thread_inst_executed_op_dfma_pred_on.sum.peak_sustained': peak_dfma = vv.tolist()
    # elif kk == 'sm__cycles_elapsed.avg.per_second'
    elif kk == 'dram__bytes.sum.peak_sustained': peak_dram_bytes = vv.tolist()
    elif kk == 'dram__cycles_elapsed.avg.per_second': peak_dram_cycpersec = vv.tolist()
    elif kk == 'device__attribute_display_name': device_name = vv.tolist()      
```

```python
len(cyc_avg)
```

```python
new_column_t = []  # Time
new_column_cc = []  # Cuda cores FLOPs
new_column_tc = []  # Tensor cores FLOPs
new_column_all = [] # All FLOPs
new_column_gf = [] # GFLOP/s
new_column_ai_hbm = []  # FLOP/byte
new_column_ai_l2 = []  # FLOP/byte
new_column_ai_l1 = []  # FLOP/byte
# new_column_peak_sp = []
new_column_peak_dp = []
new_column_peak_hbm = []
for ii in range(len(cyc_avg)):
    # --- nanoseconds = cyc : cyc/nsec
    new_column_t.append(float(cyc_avg[ii]) / float(cyc_avg_per_second[ii]))
    # --- FLOPS:
    flops = (2 * float(dfma[ii]) + float(dmul[ii]) + float(dadd[ii])) + \
            (2 * float(ffma[ii]) + float(fmul[ii]) + float(fadd[ii])) + \
            (2 * float(hfma[ii]) + float(hmul[ii]) + float(hadd[ii]))
    new_column_cc.append(flops)
    new_column_tc.append(512 * float(tensor_inst[ii]))
    new_column_all.append(new_column_cc[ii] + new_column_tc[ii])
    new_column_gf.append(new_column_all[ii] / new_column_t[ii])  # = GFLOP/s (no: 1024**3)
    # --- AI: FLOPs / BYTE
    new_column_ai_hbm.append(new_column_all[ii] / float(dram[ii]) / 1e6)
    new_column_ai_l2.append(new_column_all[ii] / float(l2[ii]) / 1e6)
    new_column_ai_l1.append(new_column_all[ii] / float(l1[ii]) / 1e6)
    # --- PEAKS:
    new_column_peak_dp.append(2 * float(peak_dfma[ii]) * float(cyc_avg_per_second[ii]))
    # new_column_peak_sp.append(2 * float(peak_ffma[ii]) * float(cyc_avg_per_second[ii])
    new_column_peak_hbm.append(float(peak_dram_bytes[ii]) * float(peak_dram_cycpersec[ii]))
    # dram__bytes.sum.peak_sustained        Kbyte/cycle     1.02
    # * dram__cycles_elapsed.avg.per_second cycle/usecond   877.5
    # = Kbyte/cycle * cycle/usecond = 895.05 Gbytes/sec
    
dfjg['Time'] = new_column_t
dfjg['CC FLOPs'] = new_column_cc
dfjg['TC FLOPs'] = new_column_tc
dfjg['all FLOPs'] = new_column_all
dfjg['GFLOP/s'] = new_column_gf
dfjg['AI HBM'] = new_column_ai_hbm
dfjg['AI L2'] = new_column_ai_l2
dfjg['AI L1'] = new_column_ai_l1
# peaks:
dfjg['Peak DP'] = new_column_peak_dp
dfjg['Peak HBM'] = new_column_peak_hbm
dfjg[['ID', 'Kernel Name', 'Time', 'CC FLOPs', 'TC FLOPs', 'all FLOPs',
      'GFLOP/s', 'AI HBM', 'AI L2', 'AI L1',
      'Peak DP', 'Peak HBM']]
# TODO: dfmetric['TC GFLOP/s'] = dfmetric['TC FLOPs'] / dfmetric['Time'] / (1024**3)
```
