```python
headers = ['x', 'y', 'z', 'idp', 'vx', 'vy', 'vz', 'rhop', 'type', 'nan']
csvfile = '/tmp/PartCsv_0049.csv'
# csvfile = '/tmp/0.csv'
df  = pd.read_csv(csvfile, sep=';', names=headers, header=1)
```

```python
print(type(df))
print(df.head(2))
```

```python
xx = df['x']
yy = df['y']
zz = df['z']
cc = df['type']
```
